```python
import numpy as np
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import f1_score
import matplotlib.pyplot as plt

# Load the dataset
data = np.loadtxt('assignment2_q2_dataset.txt')
X = data[:, :-1]  # Features
y = data[:, -1]   # Labels

# Initialize variables to store F1 scores
num_runs = 20
max_poly_order = 10
f1_train_scores = np.zeros((max_poly_order, num_runs * 5))
f1_test_scores = np.zeros((max_poly_order, num_runs * 5))

# Repeat the process 20 times
for run in range(num_runs):
    kf = KFold(n_splits=5, shuffle=True, random_state=run)
    
    for fold, (train_index, test_index) in enumerate(kf.split(X)):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        
        for order in range(1, max_poly_order + 1):
            # Generate polynomial features
            poly = PolynomialFeatures(degree=order)
            X_train_poly = poly.fit_transform(X_train)
            X_test_poly = poly.transform(X_test)
            
            # Train logistic regression model
            model = LogisticRegression(max_iter=1000)
            model.fit(X_train_poly, y_train)
            
            # Predict on training and test sets
            y_train_pred = model.predict(X_train_poly)
            y_test_pred = model.predict(X_test_poly)
            
            # Compute F1 scores
            f1_train = f1_score(y_train, y_train_pred)
            f1_test = f1_score(y_test, y_test_pred)
            
            # Store F1 scores
            f1_train_scores[order - 1, run * 5 + fold] = f1_train
            f1_test_scores[order - 1, run * 5 + fold] = f1_test

# Calculate average F1 scores across all runs and folds
avg_f1_train = np.mean(f1_train_scores, axis=1)
avg_f1_test = np.mean(f1_test_scores, axis=1)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(range(1, max_poly_order + 1), avg_f1_train, label='Training F1 Score', marker='o')
plt.plot(range(1, max_poly_order + 1), avg_f1_test, label='Test F1 Score', marker='o')
plt.xlabel('Polynomial Order')
plt.ylabel('Average F1 Score')
plt.title('Average F1 Score vs Polynomial Order')
plt.legend()
plt.grid(True)
plt.show()
print(avg_f1_train)
```
